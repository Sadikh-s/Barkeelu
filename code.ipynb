```python
import os

# Dictionnaire d'entrée (menu_data) — à définir en amont
menu_data = {
    "sectors": {
        "consumer": ["Supermarkets", "Department stores", "Fashion & apparel", "Cosmetics & beauty", "Electronics"],
        "finance": ["Banking", "Insurance", "Fintech", "Wealth management"],
        "health": ["Pharma", "Biotech", "Medtech", "Hospitals", "Health services"],
        "tech": ["Telecommunications", "IT services", "Software", "Internet", "Electronics manufacturing"],
        "transport": ["Automotive", "Rail", "Aerospace", "Urban transport"],
        "public_sec": ["Government", "Non-profit organizations", "NGOs"],
        "industry": ["Manufacturing", "Industrial equipment", "Construction", "Energy"]
    },
    "expertise": {
        "health": ["Innovative biotherapies", "Patient journey", "Rare diseases", "Digital therapeutics (DTx)",
                   "AI in health", "Personalized medicine", "Clinical trial digitalization", "HealthTech"],
        "tech": ["Usage behavior analytics", "Service optimization", "Network performance",
                 "Churn prediction", "Product-market fit"],
        "finance": ["Risk & fraud analytics", "Client segmentation", "Credit scoring",
                    "Regulatory compliance", "Fintech market intelligence"],
        "transport": ["Mobility data insights", "Predictive maintenance", "Smart infrastructure",
                      "Fleet analysis", "Demand forecasting"],
        "industry": ["Smart factory analytics", "Performance monitoring", "Supply chain optimization",
                     "Predictive quality control", "Energy efficiency"]
    },
    "services": {
        "global": ["Management", "Analysis", "Predictive modeling", "Visualization"],
        "map": ["Management", "Analysis", "Predictive modeling"],
        "mav": ["Management", "Analysis", "Visualization"],
        "mpv": ["Management", "Predictive modeling", "Visualization"],
        "apv": ["Analysis", "Predictive modeling", "Visualization"],
        "ma": ["Management", "Analysis"],
        "mv": ["Management", "Visualization"],
        "mp": ["Management", "Predictive modeling"],
        "pv": ["Predictive modeling", "Visualization"],
        "av": ["Analysis", "Visualization"],
        "ap": ["Analysis", "Predictive modeling"]
    },
    "company": {
        "blog": ["Latest news", "Expert articles", "Team stories"],
        "missions": ["Our missions", "Client stories"],
        "values": ["Transparency", "Expertise", "Diversity"]
    }
}

# Dossier de sortie
output_dir = "explore_pages"
os.makedirs(output_dir, exist_ok=True)

# Génère les fichiers HTML
for section, categories in menu_data.items():
    for cat, items in categories.items():
        for item in items:
            # Dossier
            dir_path = os.path.join(output_dir, section, cat)
            os.makedirs(dir_path, exist_ok=True)

            # Nom de fichier nettoyé
            filename = item.lower().replace(' ', '-').replace('&', 'and').replace('(', '').replace(')', '') + ".html"
            file_path = os.path.join(dir_path, filename)

            # Lien de retour vers /explore/cat
            back_link = f"/explore/{cat}"

            # HTML
            html = f"""<html>
<head>
  <title>{item}</title>
</head>
<body>
  <h1>{item}</h1>
  <p>This is the content page for <strong>{item}</strong> under <code>{section}/{cat}</code>.</p>
  <p><a href="{back_link}">← Back to Explore {cat.replace('_', ' ').title()}</a></p>
</body>
</html>"""

            with open(file_path, "w", encoding="utf-8") as f:
                f.write(html)

print(f"✅ Tous les fichiers ont été générés dans : {output_dir}/[section]/[cat]/[item].html")
```
